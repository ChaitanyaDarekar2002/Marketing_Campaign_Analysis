```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Total_Children'] = data[['Kidhome','Teenhome']].sum(axis=1)
data.head()
```

```python
from datetime import date
```

```python
current_year = date.today().year
data['Age'] = current_year - data['Year_Birth']
```

```python
data['Age'].sort_values()
```

```python
age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return('Unknown')

data['Age_Group'] = data['Age'].apply(assign_age_group)
```

```python
data[['Age', 'Age_Group']].head()
```

```python
age_order = ['23 - 30','31 - 40','41 - 50','51 - 60','61 - 70','71 - 85']
sns.pointplot(data = data, x = 'Age_Group', y = 'AcceptedCmp3', order = age_order );
```

```python
age_count = data['Age_Group'].value_counts()
```

```python
age_distribution = age_count / data.shape[0]
age_distribution = age_distribution.reset_index()
age_distribution.columns=['Age_Group', 'Percentage']
age_distribution.sort_values(by='Age_Group', inplace = True)
```

```python
sns.barplot(x='Age_Group', y='Percentage', data = age_distribution);
plt.title('Wich is the age Group that accepted most campaign by percentage?')
plt.show()
```

```python
data['MntTotal'] = data[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
```

```python
data.groupby('Age_Group')['MntTotal'].sum()
```

```python
amount_spent = data.groupby('Age_Group')['MntTotal'].sum().reset_index()
```

```python
sns.barplot(x='Age_Group', y='MntTotal', data = amount_spent, saturation= 1);
plt.title('Total Amount spent per each Age Group')
plt.show()
```

```python
acc_cmp = data[data['AcceptedCmp3'] !=0]

amount_spent_cmp = acc_cmp.groupby('Age_Group')['MntTotal'].sum().reset_index().sort_values(by='Age_Group')

sns.barplot(x='Age_Group', y='MntTotal', data = amount_spent_cmp, saturation= 1);
plt.title('Amount spent for people who have accepted campaign')
plt.show()
```

```python
purchases = pd.DataFrame(data[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(), columns = ['Sums'])
```

```python
purchases = purchases.reset_index()
```

```python
purchases.rename(columns = {'index' : 'Type_of_Purchase'}, inplace = True)
```

```python
sns.barplot(x='Type_of_Purchase', y='Sums', data = purchases, saturation= 1);
plt.title("Total Number of Purchases by Sales Channel")
plt.show()
```
